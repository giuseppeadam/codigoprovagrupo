```python
import matplotlib.pyplot as plt
x  = []
y  = []
taxa=0.03

valordoano=10000.00

for i in range (1,200):
  x.append(i+1)
  valordoano=(valordoano+(valordoano*taxa))
  y.append(valordoano)
plt.grid()
plt.plot(x,y)
plt.title('a curva ta certa mas o grafico nao')
plt.show()
```
